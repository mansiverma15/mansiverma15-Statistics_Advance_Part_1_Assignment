##Theoretical Questions

1) What is a random variable in probability theory?
- A random variable is a numerical outcome of a random experiment. It assigns a
 numerical value to each possible outcome in a sample space.

2) What are the types of random variables?
- There are two types:

 Discrete Random Variable: Takes countable values (e.g., number of heads in a coin toss).

 Continuous Random Variable: Takes uncountable values (e.g., height of a person).

3) What is the difference between discrete and continuous distributions?

- Discrete distributions deal with countable values (e.g., binomial, Poisson).

- Continuous distributions deal with uncountable values (e.g., normal,
 exponential).

4) What are probability distribution functions (PDF)?
- A Probability Distribution Function (PDF) describes the likelihood of
 different values for a continuous random variable.

5) How do cumulative distribution functions (CDF) differ from probability
 distribution functions (PDF)?

- PDF gives the probability of a specific value.

- CDF gives the cumulative probability up to a value, showing the probability of a value being less than or equal to a certain threshold.

6) What is a discrete uniform distribution?
- A discrete uniform distribution assigns equal probability to all possible
 discrete values within a finite range.

7) What are the key properties of a Bernoulli distribution?

- It has two possible outcomes: success (1) and failure (0).

 The probability of success is p, and failure is 1 - p.

 The mean is p, and variance is p(1 - p).

8) What is the binomial distribution, and how is it used in probability?

- It models the number of successes in n independent Bernoulli trials.

 Probability formula:

 𝑃(𝑋=𝑘)=(𝑛/𝑘)𝑝^𝑘(1−𝑝)^𝑛−𝑘

 Used in coin tosses, quality control, etc.

9) What is the Poisson distribution, and where is it applied?

- It models the probability of a fixed number of events occurring in a fixed
 interval.

 Formula:

 𝑃(𝑋=𝑘)=𝑒^(−𝜆)*𝜆^(𝑘)/𝑘!

 Used for rare event modeling, such as call center traffic.

10) What is a continuous uniform distribution?
- A continuous uniform distribution assigns equal probability to all values
 within a given range
 [𝑎,𝑏]
 [a,b], where the probability density function is constant.

11) What are the characteristics of a normal distribution?

- Bell-shaped and symmetric.

 Mean = Median = Mode.

 Defined by mean (𝜇) and standard deviation (𝜎).

 68-95-99.7 rule applies.

12) What is the standard normal distribution, and why is it important?

- It is a normal distribution with mean 0 and standard deviation 1.

 It allows easy comparison across datasets and is used in Z-score calculations.

13) What is the Central Limit Theorem (CLT), and why is it critical in statistics?

- CLT states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases.

 It enables statistical inference.

14) How does the Central Limit Theorem relate to the normal distribution?

- Even if the population is not normal, the sample mean distribution will be
 approximately normal for large n.

15) What is the application of Z statistics in hypothesis testing?

 Z-statistics help determine if a sample mean significantly differs from a population mean.

 Used when population variance is known.

16) How do you calculate a Z-score, and what does it represent?

 Formula:

 𝑍 =(𝑋−𝜇)/𝜎

 It measures how far a value X is from the mean μ in standard deviations.

17) What are point estimates and interval estimates in statistics?

- Point Estimate: A single value used to estimate a population parameter (e.g., sample mean).

 Interval Estimate: A range of values (e.g., confidence interval) that likely contains the population parameter.

18) What is the significance of confidence intervals in statistical analysis?

- They provide a range of likely values for a population parameter.

A 95% confidence interval means there is a 95% chance the true parameter is within the range.

19) What is the relationship between a Z-score and a confidence interval?

- Z-scores determine the critical value for confidence intervals.

 For a 95% CI, the critical Z-score is 1.96.

20) How are Z-scores used to compare different distributions?

- They standardize values, allowing comparisons across different distributions with different means and standard deviations.

21) What are the assumptions for applying the Central Limit Theorem?

- Sample size n should be sufficiently large (n ≥ 30).

 Random and independent samples.

 Finite variance.

22) What is the concept of expected value in a probability distribution?

- The expected value (mean) is the long-term average of a random variable.

 Formula:

- A probability distribution describes all possible values a random variable ccan take.

```python
# 2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

low, high = 1, 6  # Example: Rolling a fair die
x = np.arange(low, high + 1)
pmf = [1 / (high - low + 1)] * len(x)

plt.bar(x, pmf, alpha=0.7, color='blue', edgecolor='black')
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.title("PMF of a Discrete Uniform Distribution")
plt.show()
```

```python
# 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
from scipy.stats import bernoulli

def bernoulli_pdf(p, x):
    return bernoulli.pmf(x, p)

# Example usage
p = 0.5  # Probability of success
x_values = [0, 1]
probabilities = [bernoulli_pdf(p, x) for x in x_values]
print("Bernoulli Distribution PMF:", dict(zip(x_values, probabilities)))
```

```python
#18) Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

import numpy as np
from scipy.stats import norm

# Example Dataset
data = [45, 50, 55, 60, 65, 70, 75]
mean = np.mean(data)
std_dev = np.std(data)

# Given data point
x = 60

# Compute Z-score
z_score = (x - mean) / std_dev
probability = norm.cdf(z_score)

print(f"Z-score: {z_score:.2f}")
print(f"Probability (CDF): {probability:.4f}")
```

```python
#19) Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm

# Sample Data
sample_mean = 52
population_mean = 50
std_dev = 10
n = 30

# Compute Z-score
z_score = (sample_mean - population_mean) / (std_dev / np.sqrt(n))

# Compute p-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))

print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
#20) Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
from scipy.stats import t

# Sample data
data = np.array([48, 50, 52, 53, 54, 55, 57])
mean = np.mean(data)
std_err = np.std(data, ddof=1) / np.sqrt(len(data))

# Confidence Level
confidence = 0.95
df = len(data) - 1
t_critical = t.ppf((1 + confidence) / 2, df)

# Compute Confidence Interval
margin_error = t_critical * std_err
lower_bound = mean - margin_error
upper_bound = mean + margin_error

print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
#21) Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

import numpy as np
from scipy.stats import norm

# Generate Normal Data
data = np.random.normal(loc=50, scale=10, size=100)
mean = np.mean(data)
std_err = np.std(data, ddof=1) / np.sqrt(len(data))

# Confidence Interval Calculation
confidence = 0.95
z_critical = norm.ppf((1 + confidence) / 2)

# Compute Interval
margin_error = z_critical * std_err
lower_bound = mean - margin_error
upper_bound = mean + margin_error

print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
#26) Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

import numpy as np

# Generate Data
data = np.random.normal(50, 10, 1000)

# Compute Range and IQR
range_ = np.max(data) - np.min(data)
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

print(f"Range: {range_:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")
```

```python
#27) Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt

# Generate Data
data = np.random.normal(50, 15, 1000)

# Compute Z-score Normalization
mean = np.mean(data)
std_dev = np.std(data)
z_normalized_data = (data - mean) / std_dev

# Plot Before and After Transformation
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='blue', alpha=0.6)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Original Data")

plt.subplot(1, 2, 2)
plt.hist(z_normalized_data, bins=30, color='red', alpha=0.6)
plt.xlabel("Z-Score")
plt.ylabel("Frequency")
plt.title("Z-Score Normalized Data")

plt.show()
```
